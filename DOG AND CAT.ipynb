```python
data = []
dict_link_cats = r'/content/gdrive/MyDrive/dataset/DOGANDCAT/cats'
dict_link_dogs = r'/content/gdrive/MyDrive/dataset/DOGANDCAT/dogs'
newsize = (130,130)
for ima in os.listdir(dict_link_cats):
    ima_path = os.path.join(dict_link_cats,ima)
    read_ima = cv2.imread(ima_path)
    resize_ima = cv2.resize(read_ima,(130,130))
    # resize_ima = read_ima.resize(newsize)
    data.append([read_ima,str(0)])

for ima in os.listdir(dict_link_dogs):
    ima_path = os.path.join(dict_link_dogs,ima)
    read_ima = cv2.imread(ima_path)
    resize_ima = cv2.resize(read_ima,(130,130))
    # resize_ima = read_ima.resize(newsize)
    data.append([read_ima,str(1)]) 
```

```python
full_data = []
for i,k in data:
    full_data.append([i,int(k)])
    
```

```python
import random
random.shuffle(data)
data[1]
```

```python
X=[]
Y=[]

for features,labels in data:
  X.append(features)
  Y.append(labels)
```

```python
X=np.array(X)
Y=np.array(Y)
X.shape
Y.shape
```

```python
pickle.dump(X,open('X.pkl','wb'))
pickle.dump(Y,open('Y.pkl','wb'))
```

```python
import pickle
```

```python
X=pickle.load(open("X.pkl",'rb'))
Y=pickle.load(open("Y.pkl",'rb'))
```

```python
X=X/255
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=0, train_size = .80)
```

```python
X.shape
```

```python
import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D

from sklearn.model_selection import train_test_split
```

```python
model =Sequential()

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Flatten())

model.add(Dense(128,input_shape=X.shape[1:],activation='relu'))

model.add(Dense(2,activation='softmax'))

model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X,Y,epochs=70,batch_size=64)
```

```python
Y_pred = model.predict(X_test)
```

```python
model.evaluate(X_train,Y_train)
```
